```python
path = "./dataset"
```

```python
import cv2
import json
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
json_gt = [json.loads(line) for line in open(os.path.join(path, 'train_gt.json'))]
y_samples = json_gt[0]['h_samples']
```

```python
for i in range(len(json_gt)):
    gt = json_gt[i]
    gt_lanes = gt['lanes']
    raw_file = os.path.join(path, gt['raw_file'])
    img = plt.imread(raw_file)
    gt_lanes_vis = [[(x, y) for (x, y) in zip(lane, y_samples) if x >=0] for lane in gt_lanes]

    for lane in gt_lanes_vis:
        for pt in lane:
            cv2.circle(img, pt, 7, color=(0, 255, 0))

    plt.imshow(img)
    plt.show()
```
